```python
failed = []
for column, data in df.iteritems():
    if data[0] == 0.0:
        failed.append(column)

# print(failed)

```

```python
print(len(df.columns) - len(failed))

# 159 for train
print(len(failed))

# 106 for test and 106 for val 
val_cols = failed[:len(failed)//2]
test_cols = failed[len(failed)//2:]
train_df = df.drop(columns=failed)
val_df = df[val_cols]   
test_df = df[test_cols]
val_df["dt"] = df["dt"]
val_df = val_df.set_index('dt')
test_df["dt"] = df["dt"]
test_df = test_df.set_index('dt')

```

```python

train_df.to_csv("./datasets/electric_train.txt", sep='\t', index=True)
val_df.to_csv("./datasets/electric_val.txt", sep='\t', index=True)
test_df.to_csv("./datasets/electric_test.txt", sep='\t', index=True)
```

```python
def findZeros(df):
    total = len(df) * len(df.columns)
    print("Total zeros:", sum((df == 0.0).sum()))
    print("Total non-zeros:", total - sum((df == 0.0).sum()))

findZeros(train_df)
findZeros(val_df)
findZeros(test_df)
```
